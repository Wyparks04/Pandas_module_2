```python
import pandas as pd
import matplotlib.pyplot as plt

ufo_sightings = pd.read_csv("C:/Users/whitn/Documents/JanuaryProject/Pandas_module_2/scrubbed.csv", low_memory=False)

# Group by state and count sightings
state_counts = ufo_sightings["state"].value_counts()

# Plot bar graph
plt.figure(figsize=(12, 6))
state_counts.plot(kind="bar", color="yellow")

plt.xlabel("State")
plt.ylabel("Number of UFO Sightings")
plt.title("UFO Sightings by State")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by country and count sightings
country_counts = ufo_sightings["country"].value_counts()

# Plot bar graph
plt.figure(figsize=(12, 6))
country_counts.plot(kind="bar", color="lightcoral")

plt.xlabel("Country")
plt.ylabel("Number of UFO Sightings")
plt.title("UFO Sightings by Country")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count occurrences of each shape
shape_counts = ufo_sightings["shape"].value_counts()

# Plot bar graph
plt.figure(figsize=(12, 6))
shape_counts.plot(kind="bar", color="purple")

plt.xlabel("UFO Shape")
plt.ylabel("Number of Sightings")
plt.title("UFO Sightings by Shape")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert latitude and longitude to numeric format
ufo_sightings["latitude"] = pd.to_numeric(ufo_sightings["latitude"], errors="coerce")
ufo_sightings["longitude"] = pd.to_numeric(ufo_sightings["longitude "], errors="coerce")

# Remove rows with missing values
ufo_sightings = ufo_sightings.dropna(subset=["latitude", "longitude"])

# Plot scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(ufo_sightings["longitude"], ufo_sightings["latitude"], alpha=0.5, color="blue")

plt.title("UFO Sightings Across Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True, linestyle="--", alpha=0.7)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert datetime column to datetime format
ufo_sightings["datetime"] = pd.to_datetime(ufo_sightings["datetime"], errors="coerce")

# Remove rows with missing values
ufo_sightings = ufo_sightings.dropna(subset=["datetime"])

# Create scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(ufo_sightings["datetime"], range(len(ufo_sightings)), alpha=0.5, color="blue")

plt.xlabel("Date & Time")
plt.ylabel("Sightings Index")
plt.title("UFO Sightings Over Time")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.7)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count occurrences of each city
city_counts = ufo_sightings["city"].value_counts().nlargest(10)  # Top 10 cities

# Plot bar graph
plt.figure(figsize=(12, 6))
city_counts.plot(kind="bar", color="green")

plt.xlabel("City")
plt.ylabel("Number of Sightings")
plt.title("Top 10 Cities with UFO Sightings")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```
