```python
import pandas as pd
import matplotlib.pyplot as plt

# URL to the dataset
url = 'https://raw.githubusercontent.com/CodeYouOrg/DataOpenClass/main/SalaryData.csv'
df = pd.read_csv(url)

# Dropping unnecessary columns
columns_to_drop = ['Overtime_Rate', 'Incentive_Allowance', 'Other', 'ObjectId']
df = df.drop(columns=columns_to_drop)

# Dropping rows with missing values
df = df.dropna()

# Filtering data for the year 2024
filtered_data = df[df['CalYear'] == 2024]  # Corrected year filter

# Grouping by Department and summing salaries
department_total = filtered_data.groupby('Department')['YTD_Total'].sum()
department_total = department_total.sort_values(ascending=False)

# Disable scientific notation for pandas
pd.set_option('display.float_format', '{:,.0f}'.format)
print("The top 10 departments by salary spend:\n", department_total.head(10))

# Plotting the top 10 departments by salary spend
ax = department_total.head(10).plot(kind='bar')

# Correcting label placements
plt.xlabel('Department')
plt.ylabel('Total Salary')
plt.title('Total Salary by Department')

plt.show()
```
