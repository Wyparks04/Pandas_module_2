```python
import pandas as pd
import matplotlib.pyplot as plt

supermarket_data = pd.read_excel("C:/Users/whitn/Documents/JanuaryProject/Pandas_module_2/supermarket_data.xlsx")

# Convert Discount and Profit columns to numeric
supermarket_data["Discount"] = pd.to_numeric(supermarket_data["Discount"], errors="coerce")
supermarket_data["Profit"] = pd.to_numeric(supermarket_data["Profit"], errors="coerce")

# Remove rows with missing values
supermarket_data = supermarket_data.dropna(subset=["Discount", "Profit"])

# Plot scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(supermarket_data["Discount"], supermarket_data["Profit"], alpha=0.5, color="blue")

plt.title("Discount Percentage vs. Total Profit Trend")
plt.xlabel("Discount Percentage")
plt.ylabel("Total Profit")
plt.grid(True, linestyle="--", alpha=0.7)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert Profit column to numeric
supermarket_data["Profit"] = pd.to_numeric(supermarket_data["Profit"], errors="coerce")

# Get top 5 products by total profit
top_products = supermarket_data.groupby("Product Name")["Profit"].sum().nlargest(5)

# Plot bar graph
top_products.plot(kind="bar", color="skyblue")
plt.xlabel("Product Name")
plt.ylabel("Total Profit")
plt.title("Top 5 Products by Profit")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
supermarket_data = pd.read_excel("C:/Users/whitn/Documents/JanuaryProject/Pandas_module_2/supermarket_data.xlsx")

# Convert Order Date to datetime format
supermarket_data["Order Date"] = pd.to_datetime(supermarket_data["Order Date"])

# Filter data for the year 2013
sales_2013 = supermarket_data[supermarket_data["Order Date"].dt.year == 2013]

# Group by month and sum quantity ordered
monthly_orders = sales_2013.groupby(sales_2013["Order Date"].dt.month)["Quantity"].sum()

# Plot the line graph
plt.figure(figsize=(10, 6))
plt.plot(monthly_orders.index, monthly_orders.values, marker="o", linestyle="-", color="blue")

plt.title("Monthly Product Orders in 2013")
plt.xlabel("Month")
plt.ylabel("Total Products Ordered")
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the graph
plt.show()




















```
